## Creating DataFrame for all the pages

```python
# final=pd.DataFrame()
all_dfs = []

for j in range(1, 7):
    url = 'https://ambitionbox.com/list-of-companies?page={}'.format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage = requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage, 'lxml')
    company=soup.find_all('div', class_='companyCardWrapper')

    name, rating, reviews, hq, no_of_jobs = [], [], [], [], []

    for i in company:

        name.append(i.find('h2').text.strip() if i.find('h2') else np.nan)

        rating.append(i.find('div', class_='rating_text').text.strip()
                    if i.find('div', class_='rating_text') else np.nan)

        reviews.append(i.find('span', class_='companyCardWrapper__ActionCount').text.strip()
                    if i.find('span', class_='companyCardWrapper__ActionCount') else np.nan)

        hq.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip()
                if i.find('span', class_='companyCardWrapper__interLinking') else np.nan)

        no_of_jobs.append(i.find_all('span', class_='companyCardWrapper__ActionCount')[1].text.strip()
                        if len(i.find_all('span', class_='companyCardWrapper__ActionCount')) > 1 else np.nan)
            
    df = pd.DataFrame({
        'Name': name,
        'Rating': rating,
        'Reviews': reviews,
        'Head_Quarters': hq,
        'No_of_Jobs': no_of_jobs
    })

    print(f"Page {j}: {len(company)} companies")

    all_dfs.append(df)
            
final=pd.concat(all_dfs, ignore_index=True)
```
